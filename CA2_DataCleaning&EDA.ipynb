## Content before Data Cleaning and EDA

__Feature Attributes for Quantified Data:__

* ID: is a number of records in an original database. Cannot be related to the participant. It can be used for reference only.
* Age (Real) is the age of participant
* Gender: Male or Female
* Education: level of education of participant
* Country: country of origin of the participant
* Ethnicity: ethnicity of participant
* Nscore (Real) is NEO-FFI-R Neuroticism
* Escore (Real) is NEO-FFI-R Extraversion
* Oscore (Real) is NEO-FFI-R Openness to experience.
* Ascore (Real) is NEO-FFI-R Agreeableness.
* Cscore (Real) is NEO-FFI-R Conscientiousness.
* Impulsive (Real) is impulsiveness measured by BIS-11
* SS (Real) is sensation seeing measured by ImpSS
* Alcohol: alcohol consumption
* Amphet: amphetamines consumption
* Amyl: nitrite consumption
* Benzos: benzodiazepine consumption
* Caff: caffeine consumption
* Cannabis: marijuana consumption
* Choc: chocolate consumption
* Coke: cocaine consumption
* Crack: crack cocaine consumption
* Ecstasy: ecstasy consumption
* Heroin: heroin consumption
* Ketamine: ketamine consumption
* Legalh: legal highs consumption
* LSD: LSD consumption
* Meth: methadone consumption
* Mushroom: magic mushroom consumption
* Nicotine: nicotine consumption
* Semer: class of fictitious drug Semeron consumption (i.e. control)
* VSA: class of volatile substance abuse consumption

__Rating's for Drug Use:__

* CL0 Never Used
* CL1 Used over a Decade Ago
* CL2 Used in Last Decade
* CL3 Used in Last Year 59
* CL4 Used in Last Month
* CL5 Used in Last Week
* CL6 Used in Last Day

# Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Drug_Consumption.csv")
```

```python
pd.set_option('display.max_columns', None)
```

```python
display(df.head())
```

```python
df = df.drop(columns=["ID"])
```

```python
display(df[df["Semer"] != "CL0"])
```

Potentional Liars => Next step is to delete them

```python
df = df[df["Semer"] == "CL0"]
```

```python
#Check
display(df[df["Semer"] != "CL0"])
```

Drop the "Liar Check" (Semer) column, as it is no longer necessary.

```python
df = df.drop(columns=["Semer"])
```

Check for duplicates

```python
df.duplicated().sum()
```

Check for missing values

```python
df.isnull().sum().sum()
```

Rename columns

```python
df = df.rename(columns={
    "Nscore"    : "Neuroticism",
    "Escore"    : "Extraversion",
    "Oscore"    : "Openness",
    "AScore"    : "Agreeableness",
    "Cscore"    : "Conscientiousness",
    "Impulsive" : "Impulsivity",
    "SS"        : "Sensation Seeking",
    "Amphet"    : "Amphetamines",
    "Amyl"      : "Amyl Nitrite",
    "Benzos"    : "Benzodiazepines",
    "Caff"      : "Caffeine",
    "Choc"      : "Chocolate",
    "Coke"      : "Cocaine",
    "Crack"     : "Crack Cocaine",
    "Legalh"    : "Legal Highs",
    "Meth"      : "Methamphetamine",
    "Mushrooms" : "Magic Mushrooms",
    "VSA"       : "Volatile Substance Abuse",
})
```

Rename rating

```python
frequency_map = {
    "CL0": "Never Used",
    "CL1": "Over a Decade Ago",
    "CL2": "Last Decade",
    "CL3": "Last Year",
    "CL4": "Last Month",
    "CL5": "Last Week",
    "CL6": "Last Day",
}
```

```python
drug_cols = [
    "Alcohol", "Amphetamines", "Amyl Nitrite", "Benzodiazepines",
    "Caffeine", "Cannabis", "Chocolate", "Cocaine", "Crack Cocaine",
    "Ecstasy", "Heroin", "Ketamine", "Legal Highs", "LSD",
    "Methamphetamine", "Magic Mushrooms", "Nicotine",
    "Volatile Substance Abuse"
]
 
for col in drug_cols:
    df[col] = df[col].map(frequency_map)
```

Rename Gender values

```python
df["Gender"] = df["Gender"].map({"M": "Male", "F": "Female"})
```

```python
personality_cols = [
    "Neuroticism", "Extraversion", "Openness",
    "Agreeableness", "Conscientiousness", "Impulsivity", "Sensation Seeking"
]

drug_cols = [
    "Alcohol", "Amphetamines", "Amyl Nitrite", "Benzodiazepines",
    "Caffeine", "Cannabis", "Chocolate", "Cocaine", "Crack Cocaine",
    "Ecstasy", "Heroin", "Ketamine", "Legal Highs", "LSD",
    "Methamphetamine", "Magic Mushrooms", "Nicotine", "Volatile Substance Abuse"
]

freq_order = [
    "Never Used", "Over a Decade Ago", "Last Decade",
    "Last Year", "Last Month", "Last Week", "Last Day"
]
sns.set_theme(style="whitegrid", palette="muted")
print(f"Dataset shape: {df.shape}")
```

```python
never_pct = []

for drug in drug_cols:
    never = (df[drug] == "Never Used").mean() * 100
    never_pct.append(never)


usage_df = pd.DataFrame({
    "Drug": drug_cols,
    "Never %": never_pct
})


usage_df = usage_df.sort_values("Never %", ascending=False)


plt.figure(figsize=(10, 8))
plt.barh(usage_df["Drug"], usage_df["Never %"], color='blue')
plt.xlabel("Percentage (%)")
plt.title("Never Used")
plt.gca().invert_yaxis()  

for i, v in enumerate(usage_df["Never %"]):
    plt.text(v + 1, i, f"{v:.1f}%", va='center')

plt.tight_layout()
plt.show()
```

```python
for drug in drug_cols:
    plt.figure(figsize=(10, 6))
    counts = df[drug].value_counts().reindex(freq_order, fill_value=0)
    plt.bar(freq_order, counts.values, color='steelblue')

    plt.title(f"Usage Frequency: {drug}", fontsize=14, fontweight='bold')
    plt.xlabel("Usage Frequency")
    plt.ylabel("Number of People")
    plt.xticks(rotation=45)
    for i, v in enumerate(counts.values):
        if v > 0:
            plt.text(i, v + 5, str(v), ha='center', fontsize=9)
    
    plt.tight_layout()
    plt.show()
```

```python
usage_map = {
    "Never Used": 0,
    "Over a Decade Ago": 1,
    "Last Decade": 2,
    "Last Year": 3,
    "Last Month": 4,
    "Last Week": 5,
    "Last Day": 6
}

drug_numeric = pd.DataFrame()
for drug in drug_cols:
    drug_numeric[drug] = df[drug].map(usage_map)

combined_data = pd.concat([df[personality_cols], drug_numeric], axis=1)

all_correlations = combined_data.corr()

corr = all_correlations.loc[personality_cols, drug_cols]

plt.figure(figsize=(20, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdBu_r", center=0)
plt.title("Personality Traits vs Drug Usage - Correlation Heatmap", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
male_rates = []
female_rates = []

for drug in drug_cols:
    male_users = df[(df["Gender"] == "Male") & (df[drug].isin(["Last Month", "Last Week", "Last Day"]))]
    male_pct = (len(male_users) / len(df[df["Gender"] == "Male"])) * 100
    male_rates.append(male_pct)
    
    female_users = df[(df["Gender"] == "Female") & (df[drug].isin(["Last Month", "Last Week", "Last Day"]))]
    female_pct = (len(female_users) / len(df[df["Gender"] == "Female"])) * 100
    female_rates.append(female_pct)

gender_df = pd.DataFrame({
    "Drug": drug_cols,
    "Male": male_rates,
    "Female": female_rates
})

gender_df["Difference"] = (gender_df["Male"] - gender_df["Female"]).round(1)
gender_df = gender_df.sort_values("Difference", ascending=False)

print("Active Drug Usage by Gender")
print(gender_df.round(1))

plt.figure(figsize=(14, 7))
x = range(len(drug_cols))
width = 0.35

plt.bar([i - width/2 for i in x], gender_df["Male"], width, label="Male", color='blue')
plt.bar([i + width/2 for i in x], gender_df["Female"], width, label="Female", color='orange')

plt.xlabel("Drug")
plt.ylabel("Active Users (%)")
plt.title("Active Drug Usage by Gender", fontsize=14, fontweight='bold')
plt.xticks(x, gender_df["Drug"], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

age_order = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
focus_drugs = ["Alcohol", "Cannabis", "Nicotine", "Cocaine", "Ecstasy", "LSD"]

for i, drug in enumerate(focus_drugs):

    rates = []
    for age in age_order:
        age_subset = df[df["Age"] == age]
        active = age_subset[drug].isin(["Last Month", "Last Week", "Last Day"]).mean() * 100
        rates.append(active)
    
    
    axes[i].bar(age_order, rates, color='steelblue')
    axes[i].set_title(drug, fontweight='bold')
    axes[i].set_ylabel("% Active Users")
    axes[i].tick_params(axis='x', rotation=45)
    
    
    for j, v in enumerate(rates):
        if v > 0:
            axes[i].text(j, v + 1, f"{v:.1f}%", ha='center', fontsize=8)

plt.suptitle("Active Drug Usage by Age Group", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
